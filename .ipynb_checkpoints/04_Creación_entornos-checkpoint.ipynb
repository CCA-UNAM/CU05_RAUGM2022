# Creación de un entorno para su uso desde Jupyterlab

En este punto tenemos instalado **JupyterLab** en el entorno base. 

El siguiente objetivo es crear un entorno con los paquetes requeridos por un notebook descargado de internet, o bien con los paquetes para comenzar a escribir uno nuevo.


## Consultar paquetes disponibles y sus versiones

Para la instalación de paquetes ya se en el entorno actual o durante la creación de uno nuevo se pueden indicar versiones a instalar de lo contrario el administrador de paquetes tratará de instalar las más recientes.

Para buscar un paquete en los repositorios usar el subcomando **search**:

~~~bash
mamba search ncview
mamba search python
~~~

## Primer entorno pangeo-notebook xarray matplotlib

> Para crear un entorno se utiliza el subcomando  **create**  indicando el nombre con la opción **-n** y a continuación el o los paquetes a instalar

~~~bash
mamba create -n ugm-python pangeo-notebook xarray matplotlib
~~~

> Se Puede indicar versiones expecíficas a instalar durante la creación de un entorno:

~~~bash
mamba create -n ugm-python xarray=2022.10.0 matplotlib=3.6.14 pangeo-notebook=2022.10.31
~~~

*pangeo-notebook proporciona diferentes paquetes utilizados por el proyecto* https://pangeo-forge.org/

## Segundo entorno  R

Muchos paquetes dependen de una versión específica de R, el paquete r-base permite la instalación del intérprete y algunos paquetes que se requieren para su funcioanmiento.

Las versiones de r-base (R) encontradas en el canal conda-forge varian desde 3.3.2 --- 4.2.1


~~~bash
mamba  create -y -n ugm-r r-base=4.2.0
~~~

Los paquetes de R que se proporcionan el los canales llevan como prefijo r-: 

rbokeh --> r-rbokeh
dplyr ---> r-rdplyr
shiny ---> r-shiny 
ggplot2 -> r-ggplot2
tidyr ---> r-tidyr
caret ---> r-caret  
....

## Creación / eliminación de entornos

En este punto es posible crear o eliminar un entorno para su uso en jupyterlab.


### Creación 

~~~bash
mamba create -n ugm-python3.7  python=3.7 ipykernel
~~~



## Eliminación 

Para eliminar un  entorno (un directorio) se debe usar la intrucción mamba env remove, en este caso para eliminar el entorno **ugm-netcdf4**

~~~bash
mamba env remove -n ugm-netcdf4
dRemove all packages in environment /home/soporte/Documentos/11_noviembre/vallarta/envs/ugm-netcdf4:

mamba env list
# conda environments:
#

ugm-r                    /home/soporte/Documentos/11_noviembre/vallarta/envs/ugm-r

jupyter-lab --no-browser
~~~

## Compartir un entorno

Se puede compartir un entorno generando una lista de los paquetes instalados, con esta lista otro usuario puede replicar el mismo entorno y poder trabajar con un  notebook

Estando activado el entorno a compartir ejecutar el comando **mamba env export**:

~~~bash
source etc/profile.d/conda.sh 
conda activate ugm-python3.7
mamba env export --no-builds
~~~

El resultado es la lista de los paquetes y sus versiones, se encuentra en formato **yaml** (https://es.wikipedia.org/wiki/YAML) y debe ser usado con el comando  **mamba env create**

Si la lista se encuentra en el archivo raugm.yaml para generar el mismo entorno el comando es el siguiente:

~~~bash
mamba env create -f raugm.yml
~~~

*Se debe tener en cuenta que para diferentes plataformas (linux, macos, mswin) no se encuentran todos los paquetes*
