Conda es un administrador de paquetes que permite la instalación de software en directorios llamados entornos "environment" los cuales son independientes entre  sí.  Conda se incluye en 2 distribuciones:

* Anaconda
* Miniconda

**Anaconda.**
Esta distribución incluye conda y 400+ paquetes de software: Pyhton, SciPy, NumPy, entre otros.

**Miniconda.**
Incluye conda  y  40+ paquetes escenciales: Pyhton, pip, zlib, entre otros. 

## Instalación y uso en sistemas operativos GNU/Linux

Los pasos para instalar miniconda en una distribución GNU/Linux, en este ejemplo Debian 11, son:

1. Descargar el script de instalación
2. Ejecutar el script indicando el directorio destino (entorno base)
3. Cargar las variables de entorno

# mamba

`mamba` es una reimplementación en lenguaje C++ del administrador de paquetes condas.

- Descarga de datos, metadatos y paquets de forma paralela 
- Uso de libsolv, estado del arte de una biblioteca utilizada por los administrdores de paquetes de Red Hat, Fedora y OpenSUSE 
- Utiliza las mismas opciones que conda para la administración de entornos y paquetes

### Instalación de mamba en un entorno base conda existente

~~~bash
conda install mamba -n base -c conda-forge
~~~

### Instalación a partir de Mambaforge (recomendado)

Página de descarga para diferentes OS:https://github.com/conda-forge/miniforge#download

**1. Instalación**
~~~bash
curl -LO  https://github.com/conda-forge/miniforge/releases/latest/download/Miniforge3-Linux-x86_64.sh
bash Mambaforge-Linux-x86_64.sh -b -p $PWD/mamba-base
~~~


**2. Inicialización del entorno para el uso de mamba**
~~~bash
source $PWD/mamba-base/etc/profile.d/conda.sh
~~~

**3. Listado de los entornos**
~~~bash
mamba env list
# conda environments:
#
base                     $PWD/mamba-base
~~~

**4. Uso del entorno base**
~~~bash
conda activate base
~~~

**5. Cantidad de paquetes instalados en un entorno base mamba** 
~~~bash
mamba list  | wc -l 
70
~~~

**6. Espacio en disco**

## Entornos

Cuando instalamos miniforge con la opción **-p** estamos estableciendo la ubicación del **entorno base** que no es otra cosa que el directorio donde será instalado conda, mamba, python así como paquetes básicos para su uso.  

~~~bash
mamba-base/
├── bin
├── compiler_compat
├── condabin
├── conda-meta
├── envs
├── etc
├── include
├── lib
├── libexec
├── man
├── pkgs
├── sbin
├── share
├── shell
├── ssl
├── x86_64-conda_cos6-linux-gnu
└── x86_64-conda-linux-gnu
~~~

Un entorno es un **directorio** con software **autocontenido**, cualquier instalación de software adicional solo ocurre dentro de ese entorno o directorio así como todas sus dependencias.  

Al crear un entorno el directorio correspondiente es creado dentro del directorio **envs** del entorno base.

## Activación / Desactivación de un entorno

**Activar** un entorno **hace disponible** el software instalado en la sesión actual, en el caso de **GNU/Linux** agrega los directorios con archivos ejecutables a la variable **$PATH**. 

Por otro lado **desactivar** un entorno es **deshacer** los cambios ralizados en la sesión actual para tener acceso al software del entorno. 

Los cambios realizados durante activación / desactivación  de un entorno depende del sistema operativo.

## Canal de software (channel)
Un canal de software es un repositorio o en otras palabras una forma de hacer referencia al ubicación de paquetes de software. 

> Mamba utiliza por default el canal  conda-forge (https://conda-forge.org/)

>Conda utiliza por default sus propopios cananes y algunos requieren del pago de una suscripción (https://legal.anaconda.com/policies/en/?name=contracting-hub#purchased-vs-free-offerings)
